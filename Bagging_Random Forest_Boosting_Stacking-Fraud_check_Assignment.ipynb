```python
# Bagged Decision Trees for Classification
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

# importing the dataset
data = pd.read_csv('Fraud_check.csv')
data.head()
```

```python
# Creating copy of a datset
data2 = data.copy()
data2.head()
```

```python
# Renamning the data
data2 = data2.rename({'Marital.Status':'Maritial_status','Taxable.Income':'Taxable_income','City.Population':'City_population','Work.Experience':'Work_experience'},axis = 1)
data2
```

```python
categorical_features = data2.describe(include=["object",'category']).columns
categorical_features
```

```python
numerical_features = data2.describe(include=["int64","float64"]).columns
numerical_features
```

```python
# converting the catogotical data into numerical 
data2['Undergrad'] = data2['Undergrad'].map({'NO':0,'YES':1})
data2['Maritial_status'] = data2['Maritial_status'].map({'Single':1,'Divorced':0,'Married':2})
data2['Urban'] = data2['Urban'].map({'YES':1,'NO':0})
data2
```

```python
# convering the catogorical data to numeric data according to problem statement
data2["taxable_income"]="GOOD"
data2.loc[data2["Taxable_income"]<=30000,"taxable_income"]="Risky"
data2.drop(['Taxable_income'],axis = 1,inplace = True)
data2
```

```python

array = data2.values
X = array[:,0:5]
Y = array[:,5]

seed=7

# splitting the data into train and test split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.33,random_state = seed)


kfold = KFold(n_splits=10, shuffle=True, random_state=seed)
cart = DecisionTreeClassifier()
num_trees = 100
model = BaggingClassifier(base_estimator=cart, n_estimators=num_trees, random_state=seed)
results = cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
```

```python
# Random Forest Classification
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

X = array[:,0:5]
Y = array[:,5]
num_trees = 100
max_features = 3
kfold = KFold(n_splits=10, shuffle=True, random_state=7)
model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)
results = cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
```

```python
# Splitting the data on train and test 
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.33,random_state = 0)
```

```python
# bulding model 
RD = RandomForestClassifier(n_estimators=num_trees,max_features=max_features)
RD.fit(x_train,y_train)# fitting on train data
```

```python
predict = RD.predict(x_test)# predicting on data
np.mean(predict==y_test)*100
```

```python
# actual vs predicted
df_t = pd.DataFrame({'Actual':y_test, 'Predicted':predict})
df_t
```
